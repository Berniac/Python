## Problem 1: Iris Dataset

```python
iris = pd.read_csv('iris.txt', sep='\t')
ames = pd.read_csv('ames_housing.txt', sep='\t')
titanic = pd.read_csv('titanic.txt', sep='\t')
batting = pd.read_csv('batting.csv', sep=',')
```

```python
iris_means = iris.groupby(['species']).mean()
iris_means
```

```python
iris_colors = ['lightcoral', 'orange', 'darkorchid']
iris_columns = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

plt.figure(figsize=[8, 6])
for i, s in enumerate(iris_means.columns):
    plt.subplot(2, 2, i+1)
    plt.bar(iris_means.index,
            iris_means.iloc[:, i],
            edgecolor='black', color=iris_colors, label=iris_means.index)
    plt.title(f'Mean {iris_columns[i]}')
    
plt.tight_layout()
plt.show()
```

## Problem 6: Batting Data - Home Runs

```python
batting.head(5)
```

```python
batting.columns.values
```

```python
total_hr = batting.loc[:, ['teamID', 'yearID','HR']].groupby(['teamID', 'yearID'], as_index=False).sum()
total_hr.sort_values('HR',ascending=False, inplace=True)
total_hr.head(10)
```

```python
avg_hr = total_hr.loc[:, ['teamID', 'HR']].groupby('teamID').mean()
avg_hr.sort_values('HR', ascending=False, inplace=True)
avg_hr.head(10)
```

## Problem 7: Batting Data - Batting Averages

```python
batting_avg = batting.loc[:, ['teamID', 'yearID', 'H', 'AB']].groupby(['teamID', 'yearID']).sum()
batting_avg.loc[:, 'BA'] = batting_avg.loc[:, 'H'] / batting_avg.loc[:, 'AB']
batting_avg.sort_values('BA', ascending=False, inplace=True)
batting_avg.head(10)
```

## Problem 8: Batting Data - Cardinals vs. Cubs

```python
stl_batting = batting.loc[(batting['teamID'] == 'SLN') & (batting['yearID'] >= 1900), ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
stl_batting.loc[:, 'BA'] = stl_batting.loc[:, 'H'] / stl_batting.loc[:, 'AB']

chi_batting = batting.loc[(batting['teamID'] == 'CHN') & (batting['yearID'] >= 1900), ['yearID', 'H', 'AB', 'HR']].groupby('yearID').sum()
chi_batting.loc[:, 'BA'] = chi_batting.loc[:, 'H'] / chi_batting.loc[:, 'AB']
```

```python
chi_batting
```

```python
plt.figure(figsize=[12, 4])
plt.subplot(1, 2, 1)
plt.plot(stl_batting.index, stl_batting.loc[:, 'BA'], color = 'blue', label = 'Cardinals' )
plt.plot(chi_batting.index, chi_batting.loc[:, 'BA'], color = 'salmon', label = 'Cubs' )
plt.xlabel('Year')
plt.ylabel('Batting Average By Year')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(stl_batting.index, stl_batting.loc[:, 'HR'], color = 'blue', label = 'Cardinals' )
plt.plot(chi_batting.index, chi_batting.loc[:, 'HR'], color = 'greenyellow', label = 'Cubs' )
plt.xlabel('Year')
plt.ylabel('Home Runs By Year')
plt.legend()

plt.show()
```
